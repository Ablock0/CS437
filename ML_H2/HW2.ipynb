#1

When we have overlapping clusters in our K-means algorithm that means that we have our cluster means being pulled / skewed in the direction of the other cluster means. So our means will be much closer together as apposed to using the true means that exist in the natural clusters. This may pose a problem for the machine learning algorithm when it goes to predict certain points because some of the points will now belong to their true cluster because the true mean in closer than it would normally be, but this can also hurt the accuracy if a point shouldnt be in the range of a cluster but because they are closer it now gets put into the wrong cluster.

#2

If we have a situation in which we have 2 unit disks that are far enough away from each other to not overlap, and we have clusters of data points within those 2 unit disks. The logical thing to do would be create two clusters, one around each disk, respectively. However, when we start trying to create too many clusters (increasing the value of K), then we start getting a bunch of very small clusters. If there are 3 clusters within one of the unit disks than it may be very difficult for the predictive model to correctly distinguish between the three, and it will be very difficult for the model to correctly predict which cluster to place a new value in. This is a great example of why analyzing and knowing your data before you actually work with it would be a very good idea, because while more clusters may sound like a good idea at first, the data actually shows otherwise.
